## Vectorizing with Gensim: Word2Vec

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
```

```python
def preprocess(text):
    return [w for w in word_tokenize(text.lower()) if w.isalpha() and w not in stop_words]
```

```python

# Assuming your DataFrame is loaded as df with these columns:
# 'lyrics', 'wiki_content', 'lyrics_sentiments'
merged_df['combined_text'] = merged_df['genius_lyrics'].fillna('') + ' ' + \
    merged_df['lastfm_wiki_content'].fillna('') + ' ' + \
    merged_df['title'].fillna('') + ' ' + \
    merged_df['album'].fillna('')

merged_df['tokens'] = merged_df['combined_text'].apply(preprocess)
```

```python
merged_df.head(1)
```

## Train Word2Vec on all tokens

```python
# Prepare corpus
corpus = merged_df['tokens'].tolist()

# Train Word2Vec model
w2v_model = Word2Vec(sentences=corpus, vector_size=300, window=5, min_count=2, workers=4, sg=1)
w2v_model.save("word2vec_lyrics.model")
```

## Vectorize each song by Averaging Word Vectors

```python
def get_average_vector(tokens, model):
    vectors = [model.wv[w] for w in tokens if w in model.wv]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)
```

```python

merged_df['vector'] = merged_df['tokens'].apply(lambda toks: get_average_vector(toks, w2v_model))
```

## Normalize numeric features

```python
from sklearn.preprocessing import StandardScaler

# Select numeric sentiment and audio columns
numeric_features = merged_df.select_dtypes(include=['number'])
scaler = StandardScaler()
```

```python
numeric_scaled = scaler.fit_transform(numeric_features)
```

```python
numeric_scaled.shape
```

```python
# Convert list of vectors into a 2D NumPy array
text_embeddings = np.vstack(merged_df['vector'].values)
text_embeddings
```

## Combine all features

```python
combined_features = np.hstack([text_embeddings, numeric_scaled])
```

```python
combined_features
```

### Removing NaN values

```python
combined_features = np.nan_to_num(combined_features)
```

## Compute similarities

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend(song_index, top_n=5):
    # Compute cosine similarity of the song with all others
    similarities = cosine_similarity([combined_features[song_index]], combined_features)[0]

    # Get top N+1 indices (including itself at index 0), then exclude self
    top_indices = similarities.argsort()[-(top_n + 1):][::-1]
    top_indices = [i for i in top_indices if i != song_index]

    # Return the recommended songs
    return merged_df.iloc[top_indices[:top_n]][['title', 'artist', 'release_date']]
```

```python
recommend(275)
```
